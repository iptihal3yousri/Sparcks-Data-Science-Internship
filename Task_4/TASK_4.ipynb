>Bivariate Analysis

```python
plt.figure(figsize = (30,30))
df_New.corr()
sns.heatmap(df_New.corr(), annot=True)
```

```python
plt.figure(figsize = (20,9))
ax = sns.countplot(x='iyear',data=df_New)
plt.xticks(rotation=90)
plt.title('Number of terrorist activites each year')
plt.show
```

```python
fig,ax = plt.subplots(figsize=(20,9))
ax = sns.barplot(x=df_New.country_txt.value_counts()[:20].index,y = df_New.country_txt.value_counts()[:20].values)
ax.set_title('Terrorist Attacks per Country')
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x = df_New['attacktype1_txt'], order = df_New['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=60)
plt.xlabel('Method')
plt.title('Method of Attack')
plt.show()
```
